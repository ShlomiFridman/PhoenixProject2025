```python
# @title Default title text
nameStudent = ""  # @param {"type":"string"}

import ipywidgets as widgets
from IPython.display import display
import json

file_path = '/content/drive/My Drive/student.txt'

with open(file_path, 'r') as file:
    students_data = json.load(file)

students = [student['name'] for student in students_data]

dropdown = widgets.Dropdown(
    options=students,
    description="Select Student:",
    disabled=False,
)

name_text = widgets.Text(description="Name:", disabled=True)
last_name_text = widgets.Text(description="Last Name:", disabled=True)
mail_text = widgets.Text(description="Mail:", disabled=True)
courses_text = widgets.Text(description="Courses:", disabled=True)
link_text = widgets.Text(description="Link:", disabled=True)

favorite_series_text = widgets.Text(
    description="Favorite Series:",
    placeholder="Enter favorite series"
)

save_button = widgets.Button(
    description="Save",
    button_style="success"
)

output = widgets.VBox([name_text, last_name_text, mail_text, courses_text, link_text, favorite_series_text, save_button])

def on_name_selected(change):
    selected_name = change.new

    student = next((s for s in students_data if s['name'].lower() == selected_name.lower()), None)

    if student:
        name_text.value = student['name']
        last_name_text.value = student['last name']
        mail_text.value = student['mail']
        courses_text.value = ', '.join(student['courses'])
        link_text.value = student['link']

        favorite_series_text.value = student.get("favorite series", "")
    else:
        name_text.value = ""
        last_name_text.value = ""
        mail_text.value = ""
        courses_text.value = ""
        link_text.value = ""
        favorite_series_text.value = ""

def update_favorite_series(change):
    selected_name = dropdown.value
    student = next((s for s in students_data if s['name'].lower() == selected_name.lower()), None)

    if student:
        student["favorite series"] = change.new

def save_updates(b):
    with open(file_path, 'w') as file:
        json.dump(students_data, file, indent=4)
    print("Updates saved to student.txt")

dropdown.observe(on_name_selected, names='value')
favorite_series_text.observe(update_favorite_series, names='value')
save_button.on_click(save_updates)

display(dropdown, output)
```
